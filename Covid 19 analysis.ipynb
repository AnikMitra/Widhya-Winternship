```python
import pandas as pd
import matplotlib as mt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/WidhyaOrg/datasets/master/covid19.csv")
```

```python
df
```

```python
df1 = df[["Date","ConfirmedIndianNational","ConfirmedForeignNational","Cured","Deaths"]]
```

```python
df1
```

```python
df2 = df1.groupby(by="Date", sort = False)
```

```python
df3 = df2[["Date","ConfirmedIndianNational","ConfirmedForeignNational","Cured", "Deaths"]].sum()
```

```python
df3["Totalcases"] = df3.sum(axis=1)
```

```python
df3
```

```python
df3["Totalcases"].plot()
```

```python
df3[["Cured", "Deaths"]].plot()
```

```python
df3["Increaseincases"] = df3["Totalcases"].diff()
```

```python
df3
```

```python
df3["Rateofincrease"] = df3["Increaseincases"] / df3["Totalcases"]
```

```python
df3
```

```python
df4 = df3.iloc[-18:,]
```

```python
df4
```

```python
r = sum(df4["Rateofincrease"])/18
```

```python
import math
```

```python
P_t = df4.iloc[0,5] * math.exp(r * 26)
```
