```python
 df = df.fillna(df.mean())
```

```python
print(df.corr())
```

```python
x = df[["btc_market_cap", "btc_n_transactions","btc_cost_per_transaction" , "btc_cost_per_transaction_percent", "btc_hash_rate", "btc_miners_revenue", "btc_difficulty"]]
x
```

```python
y = df["btc_market_price"]
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.5)
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(x_train, y_train)
```

```python
print("Model")
print("Intercept Term")
print( reg.intercept_)
print("Coefficient Term") 
print( reg.coef_)

price_pred = reg.predict(x_test)
from sklearn import metrics
print('Mean Squared Error:', metrics.mean_squared_error(y_test, price_pred))
```
